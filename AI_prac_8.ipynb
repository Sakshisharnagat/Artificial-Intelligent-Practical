Aim:- Demonstrate the concept of Semantic network on imputed
predicate logic to represent the knowledge.

```python
import networkx as nx
import matplotlib.pyplot as plt

statements = [
    "Father(John, Mike)",
    "Mother(Alice, Mike)",
    "Brother(Mike, Mary)",
    "Sister(Mary, Mike)",
    "Father(John, Alice)"
]

G = nx.DiGraph()
for s in statements:
    p, rest = s.split("(")
    subj, obj = rest[:-1].split(",")
    G.add_edge(subj.strip(), obj.strip(), label=p)

pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G,'label'), font_color='red')
plt.axis('off')
plt.show()
```
