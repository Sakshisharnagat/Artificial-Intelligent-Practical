Aim:- To develop an AI-based chatbot that answers Frequently Asked Questions (FAQs) using Artificial Intelligence techniques.



```python
# AI FAQ Chatbot

def faq_chatbot():
    print("Welcome to AI FAQ Chatbot")
    print("Type 'exit' to stop\n")

    faq = {
        "hello": "Hello! How can I help you?",
        "what is ai": "AI stands for Artificial Intelligence.",
        "what is chatbot": "A chatbot is a program that can talk with users.",
        "what is python": "Python is a programming language.",
        "who developed python": "Python was developed by Guido van Rossum.",
        "what is machine learning": "Machine Learning is a subset of AI.",
        "bye": "Goodbye! Have a nice day."
    }

    while True:
        user = input("You: ").lower()

        if user == "exit":
            print("Bot: Chat ended.")
            break
        elif user in faq:
            print("Bot:", faq[user])
        else:
            print("Bot: Sorry, I don't understand.")

faq_chatbot()
```
